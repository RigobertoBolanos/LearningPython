#Solución

```python
#Valores reales de las matrices a evaluar
targetObjetivos=[]
#Matrices a evaluar
imagesObjetivos=[]
numMatrices0 = 0
print(range(len(digits.target)))
for i in range(len(digits.target)):
    if digits.target[i] == 0 or digits.target[i] == 3:
        targetObjetivos.append(digits.target[i])
        imagesObjetivos.append(digits.images[i])
    if digits.target[i] == 0:
        numMatrices0 = numMatrices0 + 1

numMatrices3 = len(imagesObjetivos) - numMatrices0
print(targetObjetivos)
print(imagesObjetivos)
```

```python
#Se calcula una matriz promedio con el 30% de las matrices que corresponden a los números 3 y 0
matrizPromedio0 = [[None] * len(imagesObjetivos[0][0]) for i in range(len(imagesObjetivos[0][0]))]
matrizPromedio3 = [[None] * len(imagesObjetivos[0][0]) for i in range(len(imagesObjetivos[0][0]))]
for i in range(len(imagesObjetivos[0])):
    for j in range(len(imagesObjetivos[0])):
        sumatoria0 = 0.0
        sumatoria3 = 0.0
        muestra = int(len(imagesObjetivos)*0.3)
        for k in range(muestra):
            if targetObjetivos[k] == 0:
                sumatoria0 += imagesObjetivos[k][i][j]
            else:
                 sumatoria3 += imagesObjetivos[k][i][j]
        matrizPromedio0[i][j] = sumatoria0/numMatrices0
        matrizPromedio3[i][j] = sumatoria3/numMatrices3
print(matrizPromedio0)
print(matrizPromedio3)
```

```python
aproximaciones0 = []
aproximaciones3 = []
for k in range(len(imagesObjetivos)):
    sumatoriaError0 = 0.0
    sumatoriaError3 = 0.0
    for i in range(len(imagesObjetivos[0])):
        for j in range(len(imagesObjetivos[0])):
            sumatoriaError0 += abs((matrizPromedio0[i][j] - imagesObjetivos[k][i][j]))
            sumatoriaError3 += abs((matrizPromedio3[i][j] - imagesObjetivos[k][i][j]))
    aproximaciones0.append(sumatoriaError0)
    aproximaciones3.append(sumatoriaError3)
print(aproximaciones0)
print(aproximaciones3)
```

```python
#Evaluación 0 y 3
umbralAproximacion0 = 275
umbralAproximacion3 = 278
listaConfirmacion0 = []
listaConfirmacion3 = []

for i in range(len(aproximaciones0)):
    if aproximaciones0[i] <= umbralAproximacion0:
        listaConfirmacion0.append(1)
    else: 
        listaConfirmacion0.append(0)
    if aproximaciones3[i] <= umbralAproximacion3:
        listaConfirmacion3.append(1)
    else:
        listaConfirmacion3.append(0)
        
print(listaConfirmacion0)
print(listaConfirmacion3)
```

```python
def precision(targets, resultados, numeroEvaluacion):
    vp = 0
    fp = 0
    for i in range(len(targets)):
        if targets[i] == numeroEvaluacion and resultados[i] == 1:
            vp = vp + 1
        if targets[i] == numeroEvaluacion and resultados[i] == 0:
            fp = fp + 1
        if targets[i] != numeroEvaluacion and resultados[i] == 1:  
            fp = fp + 1
    print(vp)
    print(fp)
    return vp/(vp + fp)
```

```python
precision(targetObjetivos, listaConfirmacion0, 0)
```

```python
precision(targetObjetivos, listaConfirmacion3, 3)
```

```python
#Aquí se procedera a calcular los errores de la totalidad de las matrices
aproximacionesFinales0 = []
aproximacionesFinales3 = []
for k in range(len(digits.images)):
    sumatoriaError0 = 0.0
    sumatoriaError3 = 0.0
    for i in range(len(digits.images[k][0])):
        for j in range(len(digits.images[k][0])):
            sumatoriaError0 += abs((matrizPromedio0[i][j] - digits.images[k][i][j]))
            sumatoriaError3 += abs((matrizPromedio3[i][j] - digits.images[k][i][j]))
    aproximacionesFinales0.append(sumatoriaError0)
    aproximacionesFinales3.append(sumatoriaError3)
print(aproximacionesFinales0)
print(aproximacionesFinales3)
```

```python
#Evaluación de 0 y 3 en la lista de errores
listaConfirmacionFinal0 = []
listaConfirmacionFinal3 = []

for i in range(len(aproximacionesFinales0)):
    if aproximacionesFinales0[i] <= umbralAproximacion0:
        listaConfirmacionFinal0.append(1)
    else: 
        listaConfirmacionFinal0.append(0)
    if aproximacionesFinales3[i] <= umbralAproximacion3:
        listaConfirmacionFinal3.append(1)
    else:
        listaConfirmacionFinal3.append(0)
        
print(listaConfirmacionFinal0)
print(listaConfirmacionFinal3)
```

```python
precision(digits.target, listaConfirmacionFinal0, 0)
```
